```python
import pandas as pd
import re
```

```python
##导入数据&了解基本情况
data = pd.read_csv('data_emoji_deleted.csv')
print(data.head(50))
```

```python
#定义函数，清除标点符号
def clean(text):
    if not isinstance(text, str):     # 非字符串数据跳过处理
        return text
    text = re.sub(r"(回复)?(//)?\s*@\S*?\s*(:| |$)", " ", text)  # 去除正文中的@和回复/转发中的用户名
    # text = re.sub(r"#\S+#", "", text)      # 保留话题内容
    URL_REGEX = re.compile(
        r'(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:\'".,<>?«»“”‘’]))',
        re.IGNORECASE)
    text = re.sub(URL_REGEX, "", text)       # 去除网址
    text = text.replace("转发微博", "")       # 去除无意义的词语
    text = re.sub(r"\s+", " ", text) # 合并正文中过多的空格
    return text.strip()
```

```python
#删除
data_punctuation_cleaned = data
data_punctuation_cleaned['标题'] = data_punctuation_cleaned['标题'].apply(clean)
```
