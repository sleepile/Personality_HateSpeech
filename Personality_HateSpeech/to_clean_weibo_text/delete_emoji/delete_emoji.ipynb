```python
import pandas as pd
import re
```

```python
data = pd.read_csv('weibo_2024.csv')
```

```python
print(data.head())
data.info()
```

```python
#基本处理
data1 = data.drop(columns=['标题链接','图片','图片1'])
data1.info()
print(data1.head(20))
```

```python
#广泛地删除emoji符号
def remove_emojis(text):
    # 检查是否为字符串
    if not isinstance(text, str):
        return text
    # 定义一个正则表达式，用于匹配所有表情符号
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # 表情符号
        u"\U0001F300-\U0001F5FF"  # 符号和象形文字
        u"\U0001F680-\U0001F6FF"  # 交通和地图符号
        u"\U0001F700-\U0001F77F"  # 几何图形扩展-A
        u"\U0001F780-\U0001F7FF"  # 补充符号和象形文字-A
        u"\U0001F800-\U0001F8FF"  # 补充符号和象形文字-B
        u"\U0001F900-\U0001F9FF"  # 补充符号和象形文字-C
        u"\U0001FA00-\U0001FA6F"  # 补充符号和象形文字-D
        u"\U0001FA70-\U0001FAFF"  # 补充符号和象形文字-E
        u"\U00002600-\U000026FF"  # 杂项符号
        u"\U00002700-\U000027BF"  # Dingbats
        u"\U0001F1E0-\U0001F1FF"  # 国旗
        u"\U0001F1F2-\U0001F1F4"  # 国旗
        u"\U0001F1E6-\U0001F1FF"  # 国旗
        u"\U0001F600-\U0001F636"
        u"\U0001F681-\U0001F6C5"
        u"\U0001F30D-\U0001F567"
        u"\u2600-\u26FF"          # 杂项符号
        u"\u2700-\u27BF"          # Dingbats
        u"\uFE00-\uFE0F"          # 变体选择器
        u"\u1F900-\u1F9FF"        # 补充符号和象形文字
        u" U+1F239"
        "]+", flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
#开始删除
data_emoji_deleted = data1
data_emoji_deleted['标题'] = data_emoji_deleted['标题'].apply(remove_emojis)

print(data_emoji_deleted.head(20))
```

```python
#精确地删除其他符号
def remove_specific_emoji(text, emoji_code):
    # 检查是否为字符串
    if not isinstance(text, str):
        return text
    
    # 将emoji的Unicode代码转换为正则表达式模式
    emoji_pattern = re.compile(emoji_code, flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
#存放要删除的emoji，是一个列表，存放的是字符串类型的 unicode 符号
specific_emoji=["\U0001F239"]

#删除
for emoji in specific_emoji:
    data_emoji_deleted['标题'] = data_emoji_deleted['标题'].apply(lambda x: remove_specific_emoji(x, emoji))
```
